```python
#Importing Required Libraries
import pandas as pd
import numpy as np
import scipy as sp
```

```python
#Importing Required Dataset
data=pd.read_csv("/content/sample_data/california_housing_test.csv")
```

```python
#Info About data
data.info()
```

```python
#Finding mean,mode,median
#Method 1
for col in data.columns:
  print("<----------------------------->")
  print(f"Mean of {col} is {data[col].mean()}")
  print(f"Median of {col} is {data[col].median()}")
  print(f"Mode of {col} is {data[col].mode()}")
  print(f"Variance of {col} is {data[col].var()}")
  print(f"Standard Deviation of {col} is {data[col].std()}")
  print(f'Min of {col} is {data[col].min()}')
  print(f"Max of {col} is {data[col].max()}")
  print(f"25 Percentile of {col} is {data[col].quantile(0.25)}")
  print(f"75 Percentile of {col} is {data[col].quantile(0.75)}")
  print("")
```

```python
#Method 2
#Using Numpy 
#Note mode function is not available in numpy 
for col in data.columns:
  print("<-------------------------------->")
  print(f"Mean of {col} is {np.mean(data[col])}")
  print(f"Median of {col} is {np.median(data[col])}")
  print(f"Variance of {col} is {np.var(data[col])}")
  print(f"Standard Deviation of {col} is {np.std(data[col])}")
  print(f"Min of {col} is {np.min(data[col])}")
  print(f'Max of {col} is {np.max(data[col])}')
  print(f"25 Percentile of {col} is {np.percentile(data[col],25)}")
  print(f"75 Percentile of {col} is {np.percentile(data[col],75)}")
```

```python
#Method 3
import scipy as sp
from scipy import stats
for col in data.columns:
  print("<---------------------------------->")
  print(f"Mean of {col} is {sp.mean(data[col])}")
  print(f"Median of {col} is {sp.median(data[col])}")
  print(f"Mode of {col} is {stats.mode(data[col])[0]}")
  print(f"Variance of {col} is {stats.variation(data[col])}")
  print(f"Standard Deviation of {col} is {stats.tstd(data[col])}")
  print()
```
